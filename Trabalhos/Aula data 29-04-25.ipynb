```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Parâmetros
A, C = [12.0, 6.0]

# EDO
def edo_nlinear(t, h):
    h = max(h, 0.0)
    dhdt = -C * np.sqrt(h) / A
    return dhdt

# Condição inicial
h0 = [10.0]

# Resolver
sol_nlin = solve_ivp(edo_nlinear, [0, 14], h0, t_eval=np.linspace(0, 14, 15))

# Salvar valor final
h_nlin = np.zeros((1, 4))
h_nlin[0, 0] = sol_nlin.y[0, -1]

# Imprimir conjunto de pontos (tempo, h(t))
print("Conjunto de pontos (tempo, h(t)):")
for t, h in zip(sol_nlin.t, sol_nlin.y[0]):
    print(f"t = {t:.2f}, h(t) = {h:.4f}")

# Mostrar resultado final
print("\nValor final de h(t=5):", h_nlin[0, 0])

# Plotar os pontos
plt.scatter(sol_nlin.t, sol_nlin.y[0], label='h(t)', color='blue')
plt.xlabel('Tempo (t)')
plt.ylabel('h(t)')
plt.title('Solução da EDO não-linear')
plt.legend()
plt.grid(True)
plt.show()

```
